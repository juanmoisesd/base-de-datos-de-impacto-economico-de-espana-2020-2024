```python
# Install dependencies if needed
# !pip install pandas matplotlib seaborn numpy

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.dpi'] = 120
print('Libraries loaded.')
```

## 1. Load Dataset


```python
# Download from: https://doi.org/10.5281/zenodo.18879794
# df = pd.read_csv('dataset.csv')

# Synthetic example data
np.random.seed(42)
years = list(range(2000, 2024))
keywords = ['España', 'macroeconomía', 'PIB', 'inflación']
df = pd.DataFrame({'year': years,
    **{k.replace(' ','_')[:12]: np.random.normal(50,15,len(years)) for k in keywords}})

print(f'Dataset: Base de Datos de Impacto Económico de España (2020')
print(f'DOI: 10.5281/zenodo.18879794')
print(f'Shape: {df.shape}')
df.head()
```

## 2. Summary Statistics


```python
df.describe().round(2)
```

## 3. Temporal Trends


```python
fig, ax = plt.subplots(figsize=(10,5))
for col in df.columns[1:]:
    ax.plot(df['year'], df[col], marker='o', markersize=3, label=col)
ax.set_title('Base de Datos de Impacto Económico de España (2020-2024')
ax.set_xlabel('Year')
ax.set_ylabel('Value')
ax.legend(fontsize=8)
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
